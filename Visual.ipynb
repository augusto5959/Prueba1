## Leer CSV

```python
import pandas as pd
import matplotlib 

df = pd.read_csv("data/computadores.csv")
df
```

## Computadoras mÃ¡s baratas por tienda

```python
computadorasBaratas = df.loc[df.groupby('tienda')['precio'].idxmin()]
computadorasBaratas = computadorasBaratas.sort_values(by='tienda')
computadorasBaratas
```

## Precio promedio por tienda

```python
precioPromedioTienda = df.groupby('tienda')['precio'].mean().reset_index()
precioPromedioTienda
```

##  Computadoras mÃ¡s caras por estado

```python
def top5(grupo):
    return grupo.nlargest(5, 'precio')

carasEstado = df.groupby('estado').apply(top5).reset_index(drop=True)
carasEstado
```

## DistribuciÃ³n de almacenamiento

```python
resumenAlmacenamiento = df['almacenamiento'].agg(['min', 'max', 'mean'])
resumenAlmacenamiento
```

## GrÃ¡fico de barras: Promedio de precios por marca

```python
promedioPrecioMarca = df.groupby('marca')['precio'].mean()
promedioPrecioMarca.plot(kind='bar', title='Precio promedio por marca', figsize=(10, 6))
```

## GrÃ¡fico de lÃ­neas: Tendencia de precios promedio por aÃ±o

```python
precioPromedioAnio = df.groupby('anio')['precio'].mean()
precioPromedioAnio.plot(kind='line', title='Precio promedio por aÃ±o', figsize=(10, 6), xlabel='AÃ±o', ylabel='Precio Promedio', legend=True, grid=True)
```

## GrÃ¡fico circular: ProporciÃ³n de estados de computadoras

```python
conteoEstados = df['estado'].value_counts()
conteoEstados.plot(kind='pie',title='ProporciÃ³n de estados de computadoras',autopct='%1.1f%%',figsize=(8, 8),startangle=90)
```

## GrÃ¡fico de barras: Conteo de computadoras por tienda

```python
conteoTiendas = df['tienda'].value_counts()
conteoTiendas.plot(kind='bar',title='Cantidad de computadoras por tienda',xlabel='Tienda',ylabel='Cantidad',figsize=(10, 6),color='skyblue')
```

## GrÃ¡fico de lÃ­neas: EvoluciÃ³n de la cantidad de computadoras disponibles por aÃ±o

```python
conteoAnio = df['anio'].value_counts().sort_index()  
conteoAnio.plot(kind='line',title='EvoluciÃ³n de computadoras disponibles por aÃ±o',xlabel='AÃ±o',ylabel='Cantidad de computadoras',figsize=(10, 6),marker='o',grid=True,color='green')
```
