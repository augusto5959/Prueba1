## Leer CSV

```python
import pandas as pd
import matplotlib 

df = pd.read_csv("data/computadores.csv")
df
```

## Computadoras más baratas por tienda

```python
computadorasBaratas = df.loc[df.groupby('tienda')['precio'].idxmin()]
computadorasBaratas = computadorasBaratas.sort_values(by='tienda')
computadorasBaratas
```

## Precio promedio por tienda

```python
precioPromedioTienda = df.groupby('tienda')['precio'].mean().reset_index()
precioPromedioTienda
```

##  Computadoras más caras por estado

```python
def top5(grupo):
    return grupo.nlargest(5, 'precio')

carasEstado = df.groupby('estado').apply(top5).reset_index(drop=True)
carasEstado
```

## Distribución de almacenamiento

```python
resumenAlmacenamiento = df['almacenamiento'].agg(['min', 'max', 'mean'])
resumenAlmacenamiento
```

## Gráfico de barras: Promedio de precios por marca

```python
promedioPrecioMarca = df.groupby('marca')['precio'].mean()
promedioPrecioMarca.plot(kind='bar', title='Precio promedio por marca', figsize=(10, 6))
```

## Gráfico de líneas: Tendencia de precios promedio por año

```python
precioPromedioAnio = df.groupby('anio')['precio'].mean()
precioPromedioAnio.plot(kind='line', title='Precio promedio por año', figsize=(10, 6), xlabel='Año', ylabel='Precio Promedio', legend=True, grid=True)
```

## Gráfico circular: Proporción de estados de computadoras

```python
conteoEstados = df['estado'].value_counts()
conteoEstados.plot(kind='pie',title='Proporción de estados de computadoras',autopct='%1.1f%%',figsize=(8, 8),startangle=90)
```

## Gráfico de barras: Conteo de computadoras por tienda

```python
conteoTiendas = df['tienda'].value_counts()
conteoTiendas.plot(kind='bar',title='Cantidad de computadoras por tienda',xlabel='Tienda',ylabel='Cantidad',figsize=(10, 6),color='skyblue')
```

## Gráfico de líneas: Evolución de la cantidad de computadoras disponibles por año

```python
conteoAnio = df['anio'].value_counts().sort_index()  
conteoAnio.plot(kind='line',title='Evolución de computadoras disponibles por año',xlabel='Año',ylabel='Cantidad de computadoras',figsize=(10, 6),marker='o',grid=True,color='green')
```
